```python
import seaborn as sns
import matplotlib.pyplot as plt

# Filter hanya wilayah yang mengandung 'Malang'
malang_only = Ketahanan_pangan[Ketahanan_pangan['Wilayah'].str.contains('kota Malang', case=False)]

# Buat pivot tabel: index = Wilayah Malang, columns = Tahun
pivot = malang_only.pivot_table(values='Stunting (%)', index='Wilayah', columns='Tahun')

# Visualisasi heatmap
sns.heatmap(pivot, annot=True, cmap='YlGnBu')
plt.title('Heatmap Stunting Wilayah Malang per Tahun')
plt.ylabel('Wilayah')
plt.xlabel('Tahun')
plt.show()
```
